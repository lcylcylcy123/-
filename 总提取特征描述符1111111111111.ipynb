```python
def hs1(dic):
    num=[]
    k=[]
    dic1=dic.copy()
    for x in range(1,65):
        for y in range(1,65):
            for z in range(1,65):
                if dic1[str(x)+','+str(y)+','+str(z)]==str(1):
                    lsz=[z]
                    lsy=[y]
                    m=0
                    lsx=[str(x)+','+str(y)+','+str(z)]
                    dic1[lsx[m]]=str(3)
                    while dic1[lsx[m]]==str(3):
                        q=lsx[m].split(',')
                        x=int(q[0])
                        y=int(q[1])
                        z=int(q[2])
                        if 0<y+1<=64:
                            a=x
                            b=y+1
                            c=z
                            if dic1[str(a)+','+str(b)+','+str(c)]==str(1):
                                dic1[str(a)+','+str(b)+','+str(c)]=str(3)
                                lsx.append(str(a)+','+str(b)+','+str(c))
                                lsz.append(c)
                                lsy.append(b)
                        if 0<y-1<=64:
                            a=x
                            b=y-1
                            c=z
                            if dic1[str(a)+','+str(b)+','+str(c)]==str(1):
                                dic1[str(a)+','+str(b)+','+str(c)]=str(3)
                                lsx.append(str(a)+','+str(b)+','+str(c))
                                lsz.append(c)
                                lsy.append(b)
                        if 0<z+1<=64:
                            a=x
                            b=y
                            c=z+1
                            if dic1[str(a)+','+str(b)+','+str(c)]==str(1):
                                dic1[str(a)+','+str(b)+','+str(c)]=str(3)
                                lsx.append(str(a)+','+str(b)+','+str(c))
                                lsz.append(c)
                                lsy.append(b)
                        if 0<z-1<=64:
                            a=x
                            b=y
                            c=z-1
                            if dic1[str(a)+','+str(b)+','+str(c)]==str(1):
                                dic1[str(a)+','+str(b)+','+str(c)]=str(3)
                                lsx.append(str(a)+','+str(b)+','+str(c))
                                lsz.append(c)
                                lsy.append(b)
                        m=m+1
                        if len(lsx)<m+1:
                            if len(lsx)>4:
                                d1=max(lsz)-min(lsz)+1
                                d2=max(lsy)-min(lsy)+1
                                p=d1/d2
                                k.append(p)
                                num.append(len(lsx))
                            break
    return k,num
#yz平面 dz/dy
```

```python
def hs3(dic):
    m=0
    dic3=dic.copy()
    for i in range(1,65):
        for j in range(1,65):
            for k in range(1,65):
                if dic[str(i)+','+str(j)+','+str(k)]==str(1):
                    b=k-1
                    c=k+1
                    if 0<b<=64 and dic[str(i)+','+str(j)+','+str(b)]==str(2):
                        m=m+1
                    if 0<c<=64 and dic[str(i)+','+str(j)+','+str(c)]==str(2):
                        m=m+1
    return m
#上下截面面积
```

```python
def hs4(dic):
    n=0
    dic4=dic.copy()
    for i in range(1,65):
        for j in range(1,65):
            for k in range(1,65):
                if dic[str(i)+','+str(j)+','+str(k)]==str(1):
                    b=i-1
                    c=i+1
                    d=j-1
                    e=j+1
                    if 0<b<=64 and dic[str(b)+','+str(j)+','+str(k)]==str(2):
                        n=n+1
                    if 0<c<=64 and dic[str(c)+','+str(j)+','+str(k)]==str(2):
                        n=n+1
                    if 0<d<=64 and dic[str(i)+','+str(d)+','+str(k)]==str(2):
                        n=n+1
                    if 0<e<=64 and dic[str(i)+','+str(e)+','+str(k)]==str(2):
                        n=n+1
    return n
#侧面截面面积
```

```python
def hs5(dic):
    dic5=dic.copy()
    num=0
    for x in range(1,65):
        for y in range(1,65):
            for z in range(1,65):
                if dic[str(x)+','+str(y)+','+str(z)]==str(2):
                    num=num+1
    p=num/len(dic)
    n=0
    for x in range(1,65):
        for y in range(1,65):
            m=0
            for z in range(1,65):
                if dic[str(x)+','+str(y)+','+str(z)]==str(2):
                    m=m+1
            if p<m/64:
                n=n+1
    q=n/64/64
    return q
#优势路径
```

```python
def hs6(dic):
    num=0
    dic6=dic.copy()
    for x in range(1,65):
        for y in range(1,65):
            for z in range(1,65):
                if dic[str(x)+','+str(y)+','+str(z)]==str(2):
                    num=num+1
    p=num/len(dic)
    n=0
    for x in range(1,65):
        for y in range(1,65):
            m=0
            for z in range(1,65):
                if dic[str(x)+','+str(y)+','+str(z)]==str(2):
                    m=m+1
            if m/64<p*0.5:
                n=n+1
    q=n/64/64
    return q
#最短路径
```

```python
def hs7(dic):
    n=0
    dic7=dic.copy()
    for x in range(1,65):
        for y in range(1,65):
            for z in range(1,65):
                if dic[str(x)+','+str(y)+','+str(z)]==str(1):
                    n=n+1
    n=n/64/64/64
    return n
#空气体积分数
```

```python
def hs8(dic):
    SF=0
    dic8=dic.copy()
    for x in range(1,65):
        n=0
        for y in range(1,65):
            m=0
            for z in range(1,65):
                if dic8[str(x)+','+str(y)+','+str(z)]==str(2):
                    m=m+1
            if m!=0:
                n=1/m+n
        SF=1/n+SF    
    return SF/64
```

```python
def hs9(dic):
    num=0
    dic9=dic.copy()
    for x in range(1,65):
        for y in range(1,65):
            for z in range(1,65):
                if dic[str(x)+','+str(y)+','+str(z)]==str(2):
                    num=num+1
    p=num/len(dic)
    T=0
    for x in range(1,65):
        n=0
        for y in range(1,65):
            m=0
            for z in range(1,65):
                if dic[str(x)+','+str(y)+','+str(z)]==str(2):
                    m=m+1
            n=n+(m-64*p)**2/64
        T=T+n**(1/2)
    return T/64
```

```python
def hs1(dic):
    num=[]
    k=[]
    h=[]
    dic1=dic.copy()
    for x in range(1,65):
        for y in range(1,65):
            for z in range(1,65):
                if dic1[str(x)+','+str(y)+','+str(z)]==str(1):
                    lsx=[x]
                    lsy=[y]
                    m=0
                    lsz=[str(x)+','+str(y)+','+str(z)]
                    dic1[lsz[m]]=str(3)
                    while dic1[lsz[m]]==str(3):
                        q=lsz[m].split(',')
                        x=int(q[0])
                        y=int(q[1])
                        z=int(q[2])
                        if 0<y+1<=64:
                            a=x
                            b=y+1
                            c=z
                            if dic1[str(a)+','+str(b)+','+str(c)]==str(2):
                                dic1[str(a)+','+str(b)+','+str(c)]=str(3)
                                lsz.append(str(a)+','+str(b)+','+str(c))
                                lsx.append(a)
                                lsy.append(b)
                        if 0<y-1<=64:
                            a=x
                            b=y-1
                            c=z
                            if dic1[str(a)+','+str(b)+','+str(c)]==str(2):
                                dic1[str(a)+','+str(b)+','+str(c)]=str(3)
                                lsz.append(str(a)+','+str(b)+','+str(c))
                                lsx.append(a)
                                lsy.append(b)
                        if 0<x+1<=64:
                            a=x+1
                            b=y
                            c=z
                            if dic1[str(a)+','+str(b)+','+str(c)]==str(2):
                                dic1[str(a)+','+str(b)+','+str(c)]=str(3)
                                lsz.append(str(a)+','+str(b)+','+str(c))
                                lsx.append(a)
                                lsy.append(b)
                        if 0<x-1<=64:
                            a=x-1
                            b=y
                            c=z
                            if dic1[str(a)+','+str(b)+','+str(c)]==str(2):
                                dic1[str(a)+','+str(b)+','+str(c)]=str(3)
                                lsz.append(str(a)+','+str(b)+','+str(c))
                                lsx.append(a)
                                lsy.append(b)
                        m=m+1
                        if len(lsz)<m+1:
                            if len(lsz)>4:
                                d1=max(lsx)-min(lsx)+1
                                d2=max(lsy)-min(lsy)+1
                                p=d1/d2
                                k.append(p)
                                num.append(len(lsz))
                                g=len(lsx)/d2
                                h.append(g)
                            break
    return k,num,h
#yz平面 dx/dy
```

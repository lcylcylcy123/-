```python
def hs1(dic):
    num=[]
    k=[]
    h=[]
    dic1=dic.copy()
    for x in range(1,65):
        for y in range(1,65):
            for z in range(1,65):
                if dic1[str(x)+','+str(y)+','+str(z)]==str(1):
                    lsz=[z]
                    lsy=[y]
                    m=0
                    lsx=[str(x)+','+str(y)+','+str(z)]
                    dic1[lsx[m]]=str(3)
                    while dic1[lsx[m]]==str(3):
                        q=lsx[m].split(',')
                        x=int(q[0])
                        y=int(q[1])
                        z=int(q[2])
                        if 0<y+1<=64:
                            a=x
                            b=y+1
                            c=z
                            if dic1[str(a)+','+str(b)+','+str(c)]==str(1):
                                dic1[str(a)+','+str(b)+','+str(c)]=str(3)
                                lsx.append(str(a)+','+str(b)+','+str(c))
                                lsz.append(c)
                                lsy.append(b)
                        if 0<y-1<=64:
                            a=x
                            b=y-1
                            c=z
                            if dic1[str(a)+','+str(b)+','+str(c)]==str(1):
                                dic1[str(a)+','+str(b)+','+str(c)]=str(3)
                                lsx.append(str(a)+','+str(b)+','+str(c))
                                lsz.append(c)
                                lsy.append(b)
                        if 0<z+1<=64:
                            a=x
                            b=y
                            c=z+1
                            if dic1[str(a)+','+str(b)+','+str(c)]==str(1):
                                dic1[str(a)+','+str(b)+','+str(c)]=str(3)
                                lsx.append(str(a)+','+str(b)+','+str(c))
                                lsz.append(c)
                                lsy.append(b)
                        if 0<z-1<=64:
                            a=x
                            b=y
                            c=z-1
                            if dic1[str(a)+','+str(b)+','+str(c)]==str(1):
                                dic1[str(a)+','+str(b)+','+str(c)]=str(3)
                                lsx.append(str(a)+','+str(b)+','+str(c))
                                lsz.append(c)
                                lsy.append(b)
                        m=m+1
                        if len(lsx)<m+1:
                            if len(lsx)>4:
                                d1=max(lsz)-min(lsz)+1
                                d2=max(lsy)-min(lsy)+1
                                p=d1/d2
                                k.append(p)
                                num.append(len(lsx))
                                g=len(lsx)/d2
                                h.append(g)
                            break
    return k,num,h
#yz平面 dz/dy
```

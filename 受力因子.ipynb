```python
def hs10(dic):
    n=0
    m=0
    dic10=dic.copy()
    z=64
    for x in range(1,65):
        for y in range(1,65):
            if dic[str(x)+','+str(y)+','+str(z)]==str(2):
                n=n+1
    n=n/64/64
    z=1
    for x in range(1,65):
        for y in range(1,65):
            if dic[str(x)+','+str(y)+','+str(z)]==str(2):
                m=m+1
    m=m/64/64
    return n,m
```
